### Importando as bibliotecas necessárias

```python
import numpy as np
import pandas as pd
import cv2
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

### Inicialização e análise inicial dos dados

A base de dados disponível no *Kaggle* inclui 5,863 imagens de exames Raios-X no formato JPEG. Tais imagens estão distribuídas em três pastas, referentes aos conjuntos de treino, teste e validação. Em cada pasta há ainda uma separação entre imagens cujo diagnóstico é "normal" e aquelas cujo diagnóstico é "pneumonia".

A função descrita a seguir percorre e inicializa os dados, retornando um dicionário com os conjuntos e seus respectivos rótulos. Também é feita a conversão dos rótulos de dados categóricos para dados numéricos, onde "normal" é representado como 0 e "pneumonia" como 1.

```python
def load_data(data_path):    
    data = {}
    
    splits = ['train', 'test', 'val']
    labels = ['NORMAL', 'PNEUMONIA']
    
    img_size = 150
    
    for split in splits:
        split_data = []
        splits_path = os.path.join(data_path, split)
        
        for label in labels:
            labels_path = os.path.join(splits_path, label)
            
            for file in os.listdir(labels_path):
                img = cv2.imread(os.path.join(labels_path, file), cv2.IMREAD_GRAYSCALE)
                img = cv2.resize(img, (img_size, img_size))

                num_label = 0 if label == 'NORMAL' else 1
                split_data.append((img, num_label))
        
        data[split] = np.array(split_data)
    
    return data
```

Os dados se encontram no formato ($X_i$, $y_i$), onde $X_i$ representa a $i$-ésima imagem e $y_i$ o seu respectivo rótulo.

```python
data = load_data('../data')

print("Formato do conjunto de treino: ", data['train'].shape)
print("Formato do conjunto de validação: ", data['val'].shape)
print("Formato do conjunto de teste: ", data['test'].shape)
```

A seguir, mostramos duas imagens de exemplo, em escala de cinza e com o respectivo rótulo. A primeira imagem pertence à classe "normal", e a segunda à classe "pneumonia". As imagens estão em tamanhos variados, mas foram redimensionadas para um tamanho único.

```python
img_1, label_1 = data['train'][0]
img_2, label_2 = data['train'][-1]

plt.figure()
plt.title('NORMAL' if label_1 == 0 else 'PNEUMONIA')
plt.imshow(img_1, cmap='gray')

plt.figure()
plt.title('NORMAL' if label_2 == 0 else 'PNEUMONIA')
plt.imshow(img_2, cmap='gray')
```

A próxima célula mostra a proporção entre dados com diagnóstico "normal" e "pneumonia". Como pode ser constatado, há um desbalanceamento nos dados, uma vez que há muito mais exames com diagnóstico positivo. Os experimentos realizados mostrarão o impacto desse desbalanceamento no modelo e, ao final, é proposta uma solução para o problema.

```python
classes = ['Normal' if i[1] == 0 else 'Pneumonia' for i in data['train']]

sns.countplot(classes)
```

### Avaliação do modelo e experimentos

Todos os experimentos de treino, teste e validação foram realizados utilizando a biblioteca *Pytorch*. O código presente na pasta *source* foi pensado de forma a permitir a variação de diversos hiperparamêtros, que podem ser modificados através de argumentos na linha de comando. Tais hiperparâmetros incluem:

* O modelo utilizado, podendo ser uma CNN simples ou a VGG16;
* Tamanho do *batch*;
* Otimizador, podendo ser o SGD ou ADAM;
* Função de perda, podendo ser a *Cross Entropy* ou *Negative Log Likelihood*;
* Taxa de aprendizado;
* Número de épocas.

Além disso, uma série de transformações é aplicada à base de dados visando *data augmentation*, como normalização e *random crop*.

Ao final de um experimento, são gerados:

* Um arquivo PT, que é o melhor modelo salvo durante o processo de treinamento;
* Um arquivo CSV, contendo os valores de acurácia e perda ao longo das épocas;
* Um arquivo NPY, que é uma matriz de confusão contendo o resultado das predições (segundo o melhor modelo) no conjunto de teste.

A seguir, definimos funções para plotar as curvas de acurácia e perda em função do número de épocas.

```python
results_path = '../results'

def plot_acc(file_name, title):
    df = pd.read_csv(os.path.join(results_path, file_name))

    train_acc = df['train_acc']
    val_acc = df['val_acc']

    plt.title(title)
    plt.xlabel("Number of epochs")
    plt.ylabel("Accuracy")

    plt.plot(train_acc, label='Train accuracy')
    plt.plot(val_acc, label='Val accuracy')

    plt.legend()
    plt.show()

def plot_loss(file_name, title):
    df = pd.read_csv(os.path.join(results_path, file_name))

    train_acc = df['train_loss']
    val_acc = df['val_loss']

    plt.title(title)
    plt.xlabel("Number of epochs")
    plt.ylabel("Loss")

    plt.plot(train_acc, label='Train loss')
    plt.plot(val_acc, label='Val los')

    plt.legend()
    plt.show()
```

Definimos também uma função para plotar um mapa de calor a partir dos valores contidos na matriz de confusão.

```python
def plot_confusion_matrix(file_name):
    matrix = np.load(os.path.join(results_path, file_name + '.npy'))
    per_class_acc = np.diag(matrix) / matrix.sum(1)

    print("Acurácia da classe 'normal' no teste: ", per_class_acc[0])
    print("Acurácia da classe 'pneumonia' no teste: ", per_class_acc[1])
    
    classes = ['Normal', 'Pneumonia']
    df_cm = pd.DataFrame(matrix, classes, classes)
    
    sns.heatmap(df_cm, annot=True, cmap='Blues', annot_kws={"size": 16}, fmt='g')
    plt.show()
```

#### CNN

Como experimento inicial, definimos uma rede neural convolucional simples, com seis camadas convolucionais e duas lineares, tendo a ReLU como função de ativação. Os seguintes hiperparâmetros foram escolhidos:

* Tamanho do *batch*: 16
* Otimizador: SGD
* Função de perda: *Cross Entropy*
* Taxa de aprendizado: 0.001
* Número de épocas: 10

```python
cnn = 'cnn_batch16_opSGD_losscross_entropy_lr0.001_epochs10_20201029_192755.pt'

plot_acc(cnn + '.csv', 'CNN, Learning Rate 0.001')
plot_loss(cnn + '.csv', 'CNN, Learning Rate 0.001')
```

```python
plot_confusion_matrix(cnn)
```

A seguir, aumentamos a taxa de aprendizado para 0.01:

```python
cnn_lr = 'cnn_batch16_opSGD_losscross_entropy_lr0.01_epochs10_20201029_230540.pt'

plot_acc(cnn_lr + '.csv', 'CNN, Learning Rate 0.01')
plot_loss(cnn_lr + '.csv', 'CNN, Learning Rate 0.01')
```

```python
plot_confusion_matrix(cnn_lr)
```

#### VGG16

Os próximos experimentos utilizam a VGG16, que é um modelo mais robusto e amplamente utilizado em experimentos envolvendo visão computacional. Primeiramente, mantivemos os hiperparâmetros definidos anteriormente para comparar a perfomance dos modelos.

```python
vgg16 = 'vgg16_batch16_opSGD_losscross_entropy_lr0.001_epochs10_20201028_190258.pt'

plot_acc(vgg16 + '.csv', 'VGG16, Batch Size 16, SGD, Cross Entropy')
plot_loss(vgg16 + '.csv', 'VGG16, Batch Size 16, SGD, Cross Entropy')
```

```python
plot_confusion_matrix(vgg16)
```

A seguir, utilizamos diferentes otimizadores e funções de perda no mesmo modelo. O otimizador escolhido foi o ADAM, e a função de perda é o *Negative Log Likelihood*.

```python
vgg16_adam = 'vgg16_batch16_opADAM_losslog_likelihood_lr0.001_epochs10_20201030_003119.pt'

plot_acc(vgg16_adam + '.csv', 'VGG16, Batch Size 16, ADAM, Negative Log Likelihood')
plot_loss(vgg16_adam + '.csv', 'VGG16, Batch Size 16, ADAM, Negative Log Likelihood')
```

```python
plot_confusion_matrix(vgg16_adam)
```

Por fim, aumentamos o tamanho do *batch* de 16 para 128.

```python
vgg16_128_adam = 'vgg16_batch128_opADAM_losslog_likelihood_lr0.001_epochs10_20201028_005809.pt'

plot_acc(vgg16_128_adam + '.csv', 'VGG16, Batch Size 128, ADAM, Negative Log Likelihood')
plot_loss(vgg16_128_adam + '.csv', 'VGG16, Batch Size 128, ADAM, Negative Log Likelihood')
```

```python
plot_confusion_matrix(vgg16_128_adam)
```

#### VGG16 + Weighted Random Sampler

Os experimentos anteriores mostram uma boa performance dos modelos no que diz respeito à classe "pneumonia"; contudo, a classe "normal" não apresenta bons resultados. Isso pode ser uma consequência de que, como apontado anteriormente, há um grande desbalanceio entre o número de imagens de cada classe. Para tentar contornar esse problema e como experimento final, modificamos a forma como o *data loader* escolhe amostras da base da dados, aumentando a probabilidade de que imagens da classe "normal" sejam escolhidas. Como pode ser observado na matriz de confusão, essa medida melhora significativamente a performance da classe "normal".

```python
vgg16_rnd = 'vgg16_batch128_opADAM_losslog_likelihood_lr0.001_epochs10_20201030_040925.pt'

plot_acc(vgg16_rnd + '.csv', 'VGG16, Batch Size 128, ADAM, Negative Log Likelihood')
plot_loss(vgg16_rnd + '.csv', 'VGG16, Batch Size 128, ADAM, Negative Log Likelihood')
```

```python
plot_confusion_matrix(vgg16_rnd)
```
