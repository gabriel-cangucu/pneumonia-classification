### Inicialização e análise inicial dos dados

A base de dados disponível no *Kaggle* inclui 5,863 imagens de exames Raios-X no formato JPEG. Tais imagens estão distribuídas em três pastas, referentes aos conjuntos de treino, teste e validação. Em cada pasta há ainda uma separação entre imagens cujo diagnóstico é "normal" e aquelas cujo diagnóstico é "pneumonia".

A função descrita a seguir percorre e inicializa os dados, retornando um dicionário com os conjuntos e seus respectivos rótulos. Também é feita a conversão dos rótulos de dados categóricos para dados numéricos, onde "normal" é representado como 0 e "pneumonia" como 1.

```python
def load_data(data_path):
    """
    Lê os arquivos no diretório especificado e os inicializa em um dicionário
    no formato (imagem, label)
    
    Argumentos:
    -- data_path: string indicando o caminho para os dados
    
    Retorna:
    -- data: dicionário contendo as imagens e seus respectivos labels
    """
    
    data = {}
    
    splits = ['train', 'test', 'val']
    labels = ['NORMAL', 'PNEUMONIA']
    
    img_size = 150
    
    for split in splits:
        split_data = []
        splits_path = os.path.join(data_path, split)
        
        for label in labels:
            labels_path = os.path.join(splits_path, label)
            
            for file in os.listdir(labels_path):
                img = cv2.imread(os.path.join(labels_path, file), cv2.IMREAD_GRAYSCALE)
                img = cv2.resize(img, (img_size, img_size))

                num_label = 0 if label == 'NORMAL' else 1
                split_data.append((img, num_label))
        
        data[split] = np.array(split_data)
    
    return data
```

Os dados se encontram no formato ($X_i$, $y_i$), onde $X_i$ representa a $i$-ésima imagem e $y_i$ o seu respectivo rótulo.

```python
data = load_data('../data')

print("Formato do conjunto de treino: ", data['train'].shape)
print("Formato do conjunto de validação: ", data['val'].shape)
print("Formato do conjunto de teste: ", data['test'].shape)
```

A seguir, mostramos duas imagens de exemplo, em escala de cinza e com o respectivo rótulo. A primeira imagem pertence à classe "normal", e a segunda à classe "pneumonia". As imagens estão em tamanhos variados, mas foram redimensionadas para um tamanho único.

```python
img_1, label_1 = data['train'][0]
img_2, label_2 = data['train'][-1]

plt.figure()
plt.title('NORMAL' if label_1 == 0 else 'PNEUMONIA')
plt.imshow(img_1, cmap='gray')

plt.figure()
plt.title('NORMAL' if label_2 == 0 else 'PNEUMONIA')
plt.imshow(img_2, cmap='gray')
```

A próxima célula mostra a proporção entre dados com diagnóstico "normal" e "pneumonia". Como pode ser constatado, há um desbalanceamento nos dados, uma vez que há muito mais exames com diagnóstico positivo. Para contornar esse problema, foi utilizado *data augmentation* no treinamento dos dados.
